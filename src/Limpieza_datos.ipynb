```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/Depuradoras_aguas_residuales._transpuesta.csv')
df.head()
```

```python
ind=df[["YEAR", "MONTH ", "N.º de prueba"]]
df=df.drop(columns=["YEAR", "MONTH ", "N.º de prueba"])
```

```python
ind.head()
```

```python
df.head()
```

```python
columns=df.columns
```

```python
depuradoras={}
for column in columns:
    dep,val=column.split("_")
    depuradoras[dep]=depuradoras.get(dep,[])+[val]
depuradoras
    
```

```python
dataset=pd.DataFrame()
for dep,values in depuradoras.items():
    for i in range(ind.shape[0]):
        fila={
            "YEAR":ind.loc[i,"YEAR"],
            "MONTH":ind.loc[i,"MONTH "],
            "PRUEBA":ind.loc[i,"N.º de prueba"],
            "DEPURADORA":dep,
        }
        for val in values:
            fila[val]=df.loc[i,f"{dep}_{val}"]
        dataset=dataset.append(fila,ignore_index=True)
```

```python
dataset
```

. BUT para la EDAR de Butarque

. SS para Sólidos en Suspensión (mg/L) 
. PT para Fósforo Total(mg/L)
. NT para Nitrógeno Total (mg/L)

Se han generado bastantes nulos,al haber agrupado el dataset por depuradoras,no siendo posible la eliminación de estas filas,puesto que eliminariamos muchos datos del dataset,y la muestra sería demasiado pequeña,por ellos se realiza la interpolación de los datos,que es la forma más real de obtener datos más precisos y poder trabajar con ellos.

```python
dataset_inter = dataset.interpolate(method='linear')
```

```python
dataset_inter.head()
```

```python
dataset_inter["DEPURADORA_2"]=dataset_inter["DEPURADORA"].apply(lambda x :{"VIV":0, "VAL":1, "BUT":2, "GAV":2, "SOR":1, "REJ":1, "SUR":1, "CHI":3} [x])
```

```python
dataset_inter.head()
```

Una vez realizada la interpolación se observa que aún quedan 16 datos nulos en la columna de "MONTH",estos si se eliminan,ya que 16 filas de 648,no afecta prácticamente al dataset. 

```python
dataset_inter.isna().sum()
```

```python
dataset_2=dataset_inter.dropna(axis = 0)
dataset_2
```

```python
dataset_2.isna().sum()
```

```python
month = {
    "enero":1,
    "febrero":2,
    "marzo":3,
    "abril":4,
    "mayo":5,
    "junio":6,
    "julio":7,
    "agosto":8,
    "septiembre":9,
    "octubre":10,
    "noviembre":11,
    "diciembre":12
}
```

```python
dataset_2["MONTH_2"]=dataset_2["MONTH"].apply(lambda x: month[x])
```

```python
dataset_2["DATE"]=dataset_2.apply(lambda row:"-".join(map(lambda y:str(int(y)),[row["YEAR"], row["MONTH_2"], row["PRUEBA"]])) ,axis=1)
```
